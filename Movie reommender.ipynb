# Simple popularity based model


```python
class popularity_recommender():
    # To store the recommendations 
    popularity_recommendations = None
    def create(self,train_data, user_id, item_id):
        #Get a count of user_ids for each movie as recommendation score
        train_data_grouped = train_data.groupby([item_id]).agg({user_id: 'count'}).reset_index()
        train_data_grouped.rename(columns = {'user_id': 'Score'},inplace=True)
    
        #Sort the movies based on their recommendation score
        train_data_sort = train_data_grouped.sort_values(['Score', item_id], ascending = [0,1])
    
        #Generate a recommendation rank based on their score
        train_data_sort['Rank'] = train_data_sort['Score'].rank(ascending=0, method='first')
        
        self.popularity_recommendations = train_data_sort.head()

    def recommend(self,user_id):    
        user_recommendations = self.popularity_recommendations
        
        #Add user_id column for which the recommendations are being generated
        user_recommendations['user_id'] = user_id
    
        #Bring user_id column to the front
        cols = user_recommendations.columns.tolist()
        cols = cols[-1:] + cols[:-1]
        user_recommendations = user_recommendations[cols]
        
        return user_recommendations
    
```

```python
popularity_model=popularity_recommender()
popularity_model.create(train_data,'user_id','movie title')
```

# Collaborative filtering : item similarity based model

```python
class item_similarity_recommender():
    def __init__(self):
        self.train_data = None
        self.user_id = None
        self.item_id = None
        self.cooccurence_matrix = None
        
    #Get the movies corresponding to a given user
    def get_user(self, user):
        user_data = self.train_data[self.train_data[self.user_id] == user]
        user_items = list(user_data[self.item_id].unique())
        
        return user_items
        
    #Get all the unique users for a given movie
    def get_item(self, item):
        item_data = self.train_data[self.train_data[self.item_id] == item]
        item_users = set(item_data[self.user_id].unique())
            
        return item_users
        
    #Get all the movies in the training data
    def get_all_items(self):
        all_items = list(self.train_data[self.item_id].unique())
            
        return all_items
        
    #Construct the cooccurence matrix
    def construct_cooccurence_matrix(self, user_movies, all_movies):
            
        
        #Get users for all the movies in user_movies.
        user_movies_2 = []        
        for i in range(0, len(user_movies)):
            user_movies_2.append(self.get_item(user_movies[i]))
            
        cooccurence_matrix = np.matrix(np.zeros(shape=(len(user_movies), len(all_movies))), float)
           
       
        #Calculate similarity between user_movies and all_movies in the training data
        for i in range(0,len(all_movies)):
            #Calculate users (watcher) of movie i
            movie_i = self.train_data[self.train_data[self.item_id] == all_movies[i]]
            users_i = set(movie_i[self.user_id].unique())
            
            for j in range(0,len(user_movies)):       
                    
                #Get watchers  of movie j
                users_j = user_movies_2[j]
                    
                #Calculate intersection of watchers of movies i and j
                user_intersection = users_i.intersection(users_j)
                
                #Calculate cooccurence_matrix[i,j] as Jaccard Index
                if len(user_intersection) != 0:
                    #Calculate union of watchers of movies i and j
                    user_union = users_i.union(users_j)
                    
                    cooccurence_matrix[j,i] = float(len(user_intersection))/float(len(user_union))
                else:
                    cooccurence_matrix[j,i] = 0
                    
        
        return cooccurence_matrix

    
    #Use the cooccurence matrix to make top recommendations
    def generate_top_recommendations(self, user, cooccurence_matrix, all_movies, user_movies):
        print("Non zero values in cooccurence_matrix :%d" % np.count_nonzero(cooccurence_matrix))
        
        #Calculate a weighted average of the scores in cooccurence matrix for all user movies.
        user_similarity_scores = cooccurence_matrix.sum(axis=0)/float(cooccurence_matrix.shape[0])
        user_similarity_scores = np.array(user_similarity_scores)[0].tolist()
 
        #Sort the indices of user_similarity_scores based upon their value
        sort_index = sorted(((e,i) for i,e in enumerate(list(user_similarity_scores))), reverse=True)
    
        #Create a dataframe from the following
        columns = ['user_id', 'movie', 'score', 'rank']
        d_frame= pd.DataFrame(columns=columns)
         
        #Fill the dataframe with top 10 item based recommendations
        rank = 1 
        for i in range(0,len(sort_index)):
            if ~np.isnan(sort_index[i][0]) and all_movies[sort_index[i][1]] not in user_movies and rank <= 10:    
                d_frame.loc[len(d_frame)]=[user,all_movies[sort_index[i][1]],sort_index[i][0],rank]
                rank = rank+1
        
        #Handle the case where there are no recommendations
        if d_frame.shape[0] == 0:
            print("The current user has no movie for training the item similarity based recommendation model.")
            return -1
        else:
            return d_frame
 
    #Create the item similarity based recommender system model
    def create(self, train_data, user_id, item_id):
        self.train_data = train_data
        self.user_id = user_id
        self.item_id = item_id

    #Use the item similarity based recommender system model to
    #make recommendations
    def recommend(self, user):
        
       
        #Get all the movies for this user
        user_movies= self.get_user(user)    
            
        print("No. of movies for the user: %d" % len(user_movies))
        
        #Get all the movies in the training data
       
        all_movies= self.get_all_items()
        
        print("no. of movies in the training set: %d" % len(all_movies))
         
        #Construct item cooccurence matrix of size [len(user_songs) X len(all_movies)]
        cooccurence_matrix = self.construct_cooccurence_matrix(user_movies, all_movies)
        
        #Use the cooccurence matrix to make recommendations
       
        user_recommendations = self.generate_top_recommendations(user, cooccurence_matrix, all_movies, user_movies)
                
        return user_recommendations
```
